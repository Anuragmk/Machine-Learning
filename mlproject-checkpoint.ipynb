```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_sample_image

def compress_image(input_images, n_components):
    for input_image in input_images:
        image = load_sample_image(input_image)
        data_shape = image.shape
        # Reshape the image data to a 2D array (pixels, channels)
        image_reshaped = image.reshape(data_shape[0], -1)
        pca = PCA(n_components=n_components)
        image_pca = pca.fit_transform(image_reshaped)
        image_compressed = pca.inverse_transform(image_pca)
        image_compressed = image_compressed.reshape(data_shape)

        plt.figure(figsize=(8, 4))

        plt.subplot(1, 2, 1)
        plt.imshow(image)
        plt.title('Original Image')

        plt.subplot(1, 2, 2)
        plt.imshow(image_compressed.astype(np.uint8))
        plt.title(f'Compressed Image (n_components={n_components}) for {input_image}')

        plt.show()

# List of input images
input_images = ['china.jpg', 'flower.jpg']
n_components = 10

compress_image(input_images, n_components)
```
